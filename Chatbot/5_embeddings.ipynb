```python
from langchain.embeddings import AzureOpenAIEmbeddings

embeddings_model = AzureOpenAIEmbeddings(model ='dev-text-embedding-ada-002',
                                         chunk_size=1000)
```

```python
embeddings_example = {
    '안녕하세요',
    '제 이름은 홍길동 입니다.',
    '이름은 무엇입니까?',
    '랭체인은 유용합니다.',
    'Hellow World'
}
```

```python
# 각 데이터의 유사도를 비교하는 함수
from numpy import dot
from numpy.linalg import norm
import numpy as np

def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
q = embeddings_model.embed_query('이 대화에서 언급된 이름은 무엇입니까?')
a = embeddings_model.embed_query('이 대화에서 언급된 이름은 홍길동 입니다.')
print(cos_sim(q, a))

print(cos_sim(q, embeddings_result[1]))
print(cos_sim(q, embeddings_result[3]))
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

model_name = 'BAAI/bge-small-en'
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
embeddings_example = {
    'todays is monday',
    'weather is nice today',
    "what's the probloem",
    'langchain in useful',
    'Hello World',
    'my name is morris'
}

embeddings_result = hf.embed_documents(embeddings_example)
```

```python
q = hf.embed_query('Hello? who is this')
a = hf.embed_query('hi this is harrison')

print(cos_sim(q, a))
print(cos_sim(q, embeddings_result[1]))
print(cos_sim(q, embeddings_result[5]))
```
