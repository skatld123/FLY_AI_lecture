```python
import cv2

image = cv2.imread('like_lenna.png', cv2.IMREAD_GRAYSCALE)
if image is not None:
    print("이미지를 읽어왔습니다.")
else:
    print("이미지를 읽어오지 못했습니다.")
print(f"변수 타입 : {type(image)}")
```

```python
%matplotlib inline

from matplotlib import pyplot as plt

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image_big = cv2.resize(image, dsize=None, fx=2, fy=2,)
plt.imshow(cv2.cvtColor(image_big, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image_fliped = cv2.flip(image, 0)
plt.imshow(cv2.cvtColor(image_fliped, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image_fliped = cv2.flip(image, 1)
plt.imshow(cv2.cvtColor(image_fliped, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
height, width = image.shape

matrix = cv2.getRotationMatrix2D((width/2, height/2), 30, 1)
result = cv2.warpAffine(image, matrix, (width, height), borderValue=200)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imshow(cv2.cvtColor(image[:100, :100], cv2.COLOR_BGR2RGB))
plt.show()
```

```python
croped_image = image[50:150, 50:150]
croped_image[:] = 200
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image = cv2.imread('like_lenna.png', cv2.IMREAD_GRAYSCALE)
croped_image = image[50:150, 50:150].copy()
croped_image[:] = 200
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import numpy as np
space = np.zeros((500, 1000), dtype=np.uint8)
line_color = 255
space = cv2.line(space, (100, 100), (800, 400), line_color, 3, 1)

plt.imshow(cv2.cvtColor(space, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
space = np.zeros((800, 1400), dtype=np.uint8)
line_color = 255
space = cv2.ellipse(space, (500, 300), (300, 200), 0, 90, 250, line_color, 4)

plt.imshow(cv2.cvtColor(space, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
space = np.zeros((800, 1400), dtype=np.uint8)
obj1 = np.array([[300, 500], [500, 500], [400, 600], [200, 600]])
obj2 = np.array([[600, 500], [800, 500], [700, 200]])
space = cv2.polylines(space, [obj1], True, line_color, 3)
space = cv2.fillPoly(space, [obj2], line_color, 1)

plt.imshow(cv2.cvtColor(space, cv2.COLOR_BGR2RGB))
plt.show()
```
