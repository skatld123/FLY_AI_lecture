```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = "/home/npswml/Desktop/nsh/FLY_AI_lecture/openCV/Conv_Morphology/test_image.webp"
original_image = Image.open(image_path).convert("L")

# Convert image to numpy array
image_np = np.array(original_image)

data = [0, 1, 0,                                               #    
        1, 1, 1,
        0, 1, 0]
kernel = np.array(data, np.uint8).reshape(3, 3)

# Apply morphological operations
kernel = np.ones((3, 3), np.uint8)

# Erosion
eroded_image = cv2.erode(image_np, kernel, iterations=1)

# Dilation
dilated_image = cv2.dilate(image_np, kernel, iterations=1)

# Opening (Erosion followed by Dilation)
opening_image = cv2.morphologyEx(image_np, cv2.MORPH_OPEN, kernel)

# Closing (Dilation followed by Erosion)
closing_image = cv2.morphologyEx(image_np, cv2.MORPH_CLOSE, kernel)

# Plot the results
fig, axs = plt.subplots(1, 5, figsize=(15, 10))
axs[0].imshow(image_np, cmap='gray')
axs[0].set_title('Original')
axs[0].axis('off')

axs[1].imshow(eroded_image, cmap='gray')
axs[1].set_title('Eroded')
axs[1].axis('off')

axs[2].imshow(dilated_image, cmap='gray')
axs[2].set_title('Dilated')
axs[2].axis('off')

axs[3].imshow(opening_image, cmap='gray')
axs[3].set_title('Opening')
axs[3].axis('off')

axs[4].imshow(closing_image, cmap='gray')
axs[4].set_title('Closing')
axs[4].axis('off')

plt.show()
```
