## 산점도
- scatter()를 사용  
- x, y 두 변수의 상관관계를 확인하거나, 전체적인 데이터의 분포를 확인


```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.show()
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
area = (30 * np.random.rand(n))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.7)
plt.show()
```

## 막대그래프  
- 범주형 데이터의 값을 막대의 높이와 색상으로 표현  
- bar()로 그림

```python
name = ['A', 'B', 'C', 'D']
y = [5., 25., 50., 20.]
z = [8., 20., 25., 6.]

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(10, 5)
ax[0].bar(name, y, color='#33aaff')
ax[1].bar(name, z, color=['r', 'g', 'b', 'k'])
plt.show()
```

### 박스플롯  
- boxplot()으로 사용
- 데이터DML 4분위수를 표현  
- 최솟값, 중앙값, 최댓값이 나옴, 그 밖의 범위는 outlier

```python
x = np.random.normal(10, 3, 100)
y = np.random.normal(13, 3, 100)
z = np.random.normal(8, 5, 100)
plt.boxplot([x, y, z])
plt.xticks([1, 2, 3], ['A', 'B', 'C'])
plt.show()
```

### Seaborn 시각화

```python
import seaborn as sns
import pandas as pd
print(sns.__version__)
```

```python
tips = pd.read_csv('./tips.csv')
tips.head(10)
```

```python
sns.countplot(data=tips, x='day')
plt.show()
```

```python
x = tips['day'].value_counts()
print(x)
```

```python
sns.countplot(data=tips, x='day', hue='smoker')
plt.show()
```

```python
x = tips.groupby(by='sex')['tip'].mean()
print(x)
```

```python
sns.histplot(data=tips, x='total_bill', kde=True)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', kde=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x = tips['sex'], y = tips['tip'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.violinplot(x='smoker', y='total_bill', hue='smoker', data=tips)
plt.show()
```

### 타이타닉 데이터 시각화해보기

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
print(titanic.head())
print('\n')
print(titanic.info())
```

```python
sns.displot(x='pclass', data=titanic, col='survived')
plt.show()
```

```python
df_grp = titanic.groupby(['pclass', 'survived']).size().reset_index(name='count')
print(type(df_grp))
sns.barplot(x='pclass', y='count', hue='survived', data=df_grp)
plt.show()
```

```python
sns.displot(x='age', data=titanic, col='survived')
plt.show()
```

```python
sns.displot(x='age', data=titanic, col='pclass', hue='survived', legend='survived')
plt.show()
```

```python
g = sns.displot(x='age', data=titanic, col='pclass')
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="survived")
g.map(sns.scatterplot, "age", "fare")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.violinplot(x='pclass', y='age', hue='pclass', data=titanic)
plt.show()
```
